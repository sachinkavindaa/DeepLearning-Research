```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/accuracy- Sheet1.csv')
```

```python
df.columns
```

```python
plt.plot(range(1,6), df['Adam'], label = 'Adam' ,marker = '.', markersize = 10 , color = 'b')
plt.plot(range(1,6), df['SGD'], label = 'SGD' ,marker = '.', markersize = 10 , color = 'g')
plt.plot(range(1,6), df['Rmsprop'], label = 'Rmsprop' ,marker = '.', markersize = 10, color = 'r')
plt.plot(range(1,6), df['Adagrad'], label = 'Adagrad' ,marker = '.', markersize = 10, color = 'orange')
plt.title('Comparison of mean Accuracy amongst Optimizers')
plt.xlabel('Folds')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
